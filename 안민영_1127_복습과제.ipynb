행 단위 추출시 꼭 알아야 하는 내용
- loc: 인덱스를 기주능로 행 데이터 추출
-iloc: 행 번호를 기준으로 행 데이터 추출

In Pandas, "loc" is a label-based indexer for dataframes. It is used to select rows and columns from a dataframe based on their label. On the other hand, "iloc" is an integer-based indexer. It is used to select rows and columns from a dataframe based on their integer indices.

```python
df = pd.read_csv('gapminder.tsv',sep='Wt')
```

```python
df
```

```python
df.loc[1703]
```

```python
mpg.loc[233]
```

loc을 이용해서 인덱스 기준 데이터를 추출 시 변수 입력도 가능

```python
number_of_rows =mpg.shape[0]
```

```python
number_of_index=number_of_rows-12
```

```python
number_of_index
```

```python
mpg
```

```python
mpg.loc[number_of_index]
```

```python
mpg.loc[:number_of_index]
```

```python
mpg.loc[[1,10,100]]
```

iloc

```python
mpg.iloc[1]
```

```python
mpg.loc[1]
```
