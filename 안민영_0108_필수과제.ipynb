
### 재무제표 데이터 분석, 시각화

```python
import matplotlib.pyplot as plt
import pandas as pd

# Create a DataFrame from the data
df = pd.DataFrame({'Year': [2016, 2017, 2018, 2019, 2020], 
                   'Assets': [97760680817, 1.04E+11, 1.26E+11, 1.36E+11, 1.45E+11], #자산총계
                   'Liabilities': [23419492827, 23398872100, 24027374437, 34941404675, 35459747467], #부채총계
                   'Capital': [74341187990, 80950349908, 1.02E+11, 1.01E+11, 1.09E+11], #자본총계
                   'Revenue': [94993116818, 1.21E+11, 1.63E+11, 1.92E+11, 2.66E+11], #매출액
                   'Operating Income':[2379822965, 6987071352, 6412992156, 3664066464, 7227983716], #영업이익
                   'Net Income':[480426339, 7797097354, 4153603488, 3181392223, 6501208590], #당기순이익
                   'Debt-to-capital':[76.04405715, 77.57638088, 80.86578073, 74.33403127, 75.49264436]}) #부채비율

# Revenue over time
plt.plot(df["Year"],df["Revenue"])
plt.xlabel('Year')
plt.ylabel('Revenue')
plt.title('Revenue over time')
plt.show()

# Operating margin #영업이익률
df["Operating_margin"]=df["Operating Income"]/df["Revenue"]*100
plt.plot(df["Year"],df["Operating_margin"])
plt.xlabel('Year')
plt.ylabel('Operating margin')
plt.title('Operating margin')
plt.show()

# debt-to-capital ratio over time
plt.plot(df["Year"],df["Debt-to-capital"])
plt.xlabel('Year')
plt.ylabel('Debt-to-capital ratio')
plt.title('Debt-to-capital ratio over time')
plt.show()
```

```python
import numpy as np
# Bar chart of net income over time
plt.bar(df["Year"],df["Net Income"])
plt.xlabel('Year')
plt.ylabel('Net Income')
plt.title('Net Income over time')
plt.show()

# Scatter plot of Assets vs Liabilities
plt.scatter(df["Assets"], df["Liabilities"])
plt.xlabel('Assets')
plt.ylabel('Liabilities')
plt.title('Assets vs Liabilities')
plt.show()

# stacked bar chart of Assets, Liabilities and Capital over time
barWidth = 0.25
bar1 = df["Assets"]
bar2 = df["Liabilities"]
bar3 = df["Capital"]
r1 = np.arange(len(bar1))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
plt.bar(r1, bar1, color='#7f6d5f', width=barWidth, edgecolor='white', label='Assets')
plt.bar(r2, bar2, color='#557f2d', width=barWidth, edgecolor='white', label='Liabilities')
plt.bar(r3, bar3, color='#2d7f5e', width=barWidth, edgecolor='white', label='Capital')
plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Assets, Liabilities, and Capital over time')
plt.xticks([r + barWidth for r in range(len(bar1))], df["Year"])
plt.legend()
plt.show()
```

```python
import seaborn as sns
```

```python
# Histogram of Operating margin
plt.hist(df["Operating_margin"], bins = 10, color = "blue", edgecolor = "black")
plt.xlabel("Operating margin")
plt.ylabel("Frequency")
plt.title("Distribution of Operating margin")
plt.show()

# Correlation heatmap
plt.figure(figsize=(10,5))
corr = df.corr()#default값이 피어슨 상관계수임
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, annot=True)
plt.title("Correlation Heatmap")
plt.show()

# Pivot table of Operating Income by Year
pivot = df.pivot_table(values="Operating Income", index="Year")
pivot.plot(kind="bar", color = "blue")
plt.xlabel("Year")
plt.ylabel("Operating Income")
plt.title("Operating Income by Year")
plt.show()
```
