```python
# read the csv file into a pandas DataFrame
df = pd.read_csv('mpg.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
#groupby 활용해보기
```

```python
df.groupby('category')['cty'].mean()
```

```python
df.groupby('manufacturer')['cty'].mean()
```

```python
df.groupby('model')[['cty','hwy']].mean()
```

```python
df.groupby('model').agg({'cty':'mean','hwy':'mean'})
```

```python

# group the dataset by 'manufacturer' column and get the mean and max of 'mpg' and 'horsepower'
grouped = df.groupby('manufacturer').agg({'cty': ['mean', 'max'], 'hwy': ['mean', 'max']})
grouped
```
