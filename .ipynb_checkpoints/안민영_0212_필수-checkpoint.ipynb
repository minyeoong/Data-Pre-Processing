### JP,UK의 UnitPrice의 정규분포 시각화

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

```python
JP_pdf = stats.norm.pdf(df_JP["UnitPrice"].sort_values(), JP_mean, JP_std)
UK_pdf = stats.norm.pdf(df_UK["UnitPrice"].sort_values(), UK_mean, UK_std)
```

```python
plt.plot(df_JP["UnitPrice"].sort_values(), JP_pdf, color="Black", label="JP UnitPrice")
plt.plot(df_UK["UnitPrice"].sort_values(), UK_pdf, color="Orange", label="UK UnitPrice")
plt.xlim([-30,30])
plt.ylim([0,0.23])
plt.legend()
plt.xlabel("UnitPrice", size=15)
plt.ylabel("Frequency", size=15)
plt.grid(True, alpha=0.3, linestyle="--")
plt.rcParams["figure.figsize"] = [7,5]  # 가로, 세로 인치 조정
plt.rcParams["figure.dpi"] = 500   # 해상도 조정
plt.show()
```

```python
def monthly_unit_price_top_n_customers(df, n=10):
    # CustomerID를 groupby해준다. InvoiceDate 기준 ,월간
    grouped = df.groupby(['CustomerID', pd.Grouper(key='InvoiceDate', freq='M')])
    # 각 ID별 전체 UnitPrice의 sum을 구하고 내림차순으로 정렬
    total_unit_price = grouped['UnitPrice'].sum().groupby('CustomerID').sum().sort_values(ascending=False)

    print(f'Top {n} Customers by Total Unit Price:')
    print(total_unit_price.iloc[:n])
```

```python
monthly_unit_price_top_n_customers(df,5)
```

### 같은 코드를 매번 작성하지 않기 위해 CustoemrID를 매개변수로 데이터를 추출하는 함수 정의

```python
extract_customer_data(df,14911)
```

```python

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Select the relevant features for clustering
X = df[['Quantity', 'UnitPrice']]

# Scale the data using StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply K-means clustering with 5 clusters
kmeans = KMeans(n_clusters=5, random_state=0).fit(X_scaled)

# Get the cluster labels for each data point
labels = kmeans.labels_

# Add the cluster labels to the original DataFrame
df['Cluster'] = labels
```

```python
#0번째 Cluster, UnitPrice와 Quantity가 일반적인 범주에 있음
df[df['Cluster']==0]
```

```python
#1번째 Cluster, Quantity는 0근처이고 (-1,1 ..) UnitPrice는 매우 큼
df[df['Cluster']==1]
```

```python
#2번째 Cluster, Quantity가 매우 크고 UnitPrice는 매우 작음
df[df['Cluster']==2]
```

```python
#3번째 Cluster, Quantity가 음수로 매우 작고 UnitPrice는 매우 작음
df[df['Cluster']==3]
```

```python
#4번째 Cluster, Quantity가 음수이며 UnitPrice가 압도적으로 큼
df[df['Cluster']==4]
```

```python
X = df[['Quantity', 'UnitPrice']]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

#5-means적용
kmeans = KMeans(n_clusters=5, random_state=0).fit(X_scaled)

labels = kmeans.labels_

df['Cluster'] = labels

# scatterplot 시각화
plt.scatter(df['Quantity'], df['UnitPrice'], c=df['Cluster'], cmap='viridis')
plt.xlabel('Quantity')
plt.ylabel('Unit Price')
plt.title('Sales Data Clustering')
plt.show()
```
