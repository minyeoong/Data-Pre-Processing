# 데이터 시각화

-matplot,seaborn 두 개 정도로 대부분 사용

-시각화는 목적이 있어야지만 의미가 있다.

-시각화는 나의 가설, 내가 설득하고 싶을 때 사용하는 것

-시각화는 나의 가설, 나의 주장, 나의 설득과 잘 연결되고 그것을 잘 표현해야 한다.

```python
#그래프를 그리는 기초
from matplotlib import pyplot as plt
```

```python
#산점도 그리기
plt.scatter([1,2,3],[4,5,6]) #x좌표, y좌표
plt.show()
```

```python
#산점도를 그려보자!
plt.scatter([1,2,3],[4,5,6])
plt.scatter([1.5,7],[8,9])
plt.show()
```

```python
plt.scatter([1,2,3],[4,5,6])
plt.title('Hi')
plt.xlabel('x axis')
plt.ylabel('y axis')
```

```python
plt.scatter([0.5,1,1.5],[4,4.5,5])
plt.title('Title') # 제목
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.xlim(0,2)
plt.ylim(4,5)
```

```python
plt.scatter([1,2,3],[4,5,6])
plt.title('Title') # 제목
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.xlim(left = 0.25) #왼쪽 기준으로 보여주기!
plt.ylim(top=5)
```

```python
# Create a DataFrame
df = pd.DataFrame({'x': [1, 2, 3], 'y': [2, 3, 4]})

# Plot the DataFrame
df.plot(kind='line')

# Set the x-axis tick labels
plt.xticks([1, 2, 3])

# Show the plot
plt.show()
```

```python
plt.xticks([0, 1, 2], ['one', 'two', 'three'])
```

```python
plt.scatter([1,2,3],[4,5,6], label= 'x=1') #라벨 설정
plt.scatter([-1,-1,-1],[4,5,6], label= 'x=-1')
plt.legend()#라벨 출력
```

```python
plt.scatter([1,2,3],[4,5,6], label= 'x=1') #라벨 설정
plt.scatter([-1,-1,-1],[4,5,6], label= 'x=-1')
plt.legend(loc = 'upper center')#라벨 출력
```

```python
plt.figure(figsize=(10,5))
plt.scatter([1,2,3],[4,5,6])
```

```python
fig, axs = plt.subplots(2, figsize=(10,5))
axs[0].plot(df['x'],df['y'])
axs[1].plot(df['x'],df['y'])
```

```python
##스타일 바꾸기
import seaborn as sns 
sns.set()#기본값은 (style='darkgrid')
plt.scatter([1,2,3],[4,5,6])
```

```python
sns.set(style ='whitegrid')#기본값은 (style='darkgrid')
plt.scatter([1,2,3],[4,5,6])
```

```python
plt.plot([1,2,3],[4,5,6], label= 'line1')
plt.plot([1,2,3],[7,8,9], label= 'line2')
plt.legend()
```
