```python

df['일시'] = pd.to_datetime(df['일시'])

# Plot the temperature and humidity data
plt.plot(df['일시'], df['기온(°C)'], label='Temperature')
plt.plot(df['일시'], df['습도(%)'], label='Humidity')

# Add labels and title to the plot
plt.xlabel('Time')
plt.ylabel('Temperature (°C) / Humidity (%)')
plt.title('Temperature and Humidity over Time')

# Show the legend
plt.legend()

# Show the plot
plt.show()
```
